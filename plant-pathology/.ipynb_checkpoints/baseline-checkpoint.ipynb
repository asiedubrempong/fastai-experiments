## Create Datasets and DataLoaders

```python
df = pd.read_csv(path/'train.csv')
df.head()
```

Select one row and convert one hot labels to class names

```python
b  = a.loc[a ==0]
```

```python
list(b.index)
```

```python
list(a.loc[a==0].index)
```

Create datablock

```python
db = DataBlock()
```

```python
ds = db.datasets(df)
```

```python
ds.train[0]
```

```python
def get_x(r): return path/'images'/f'{r["image_id"]}.jpg'
def get_y(r): return list(r.loc[r==1].index)

db = DataBlock(get_x=get_x,
               get_y=get_y)
```

```python
ds = db.datasets(df)
```

```python
ds.train[0]
```

Good so now we're getting the right path and labels, let's specify that we're are working with images

```python
db = DataBlock(blocks=(ImageBlock, MultiCategoryBlock),
               get_x=get_x,
               get_y=get_y)
```

```python
ds = db.datasets(df)
ds.train[0]
```

```python
def get_dls(size, bs):
    db = DataBlock(blocks=(ImageBlock, MultiCategoryBlock),
                   get_x=get_x,
                   get_y=get_y,
                   splitter=RandomSplitter(seed=42),
                   item_tfms=Resize(460),
                   batch_tfms=aug_transforms(size=size, min_scale=0.75))
    return db.dataloaders(df, bs=bs)
```

```python
dls = get_dls(224, 64)
```

Let's have a look at the data

```python
dls.show_batch(max_n=9)
```

## Build baseline model

```python
learn = cnn_learner(dls, resnet18,
                    metrics=partial(accuracy_multi, thresh=0.6))
```

```python
learn.loss_func
```

```python
learn.fine_tune(10, freeze_epochs=3)
```

Find optimal threshold

```python
preds, targs = learn.get_preds()
```

```python
preds[0]
```

```python
targs[0]
```

```python
x = torch.linspace(0.05, 0.95)
```

```python
accs = [accuracy_multi(preds, targs, thresh=o)
        for o in x]
```

```python
plt.plot(x, accs)
```

```python
learn.recorder.plot_loss()
```

## Evaluate on test set

```python
test_df = pd.read_csv(path/'test.csv')
test_df.head()
```

```python
test_dl = learn.dls.test_dl(test_df, with_labels=False)
```

```python
probs, classes = learn.get_preds(dl=test_dl)
```

```python
probs.shape
```

```python
probs[0]
```

```python
probs_df = pd.DataFrame(probs)
```
